```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

location = "datasets/DC_Properties.csv"

dcprop_df = pd.read_csv(location)

print(dcprop_df.columns)
```

```python
#Dropping unwanted columns
dcprop_df = dcprop_df.drop(['QUALIFIED', 'SALE_NUM','BLDG_NUM','STRUCT','USECODE','GIS_LAST_MOD_DTTM', 'CMPLX_NUM','LIVING_GBA','NATIONALGRID','LATITUDE','LONGITUDE','ASSESSMENT_NBHD','ASSESSMENT_SUBNBHD','CENSUS_TRACT','CENSUS_BLOCK','WARD','SQUARE','X','Y'], axis=1)
```

```python
#Drop rows where price is null
dcprop_df.dropna(subset=['PRICE'])
result=dcprop_df.sort(['PRICE', 'BEDRM'], axis=1, ascending=False)
```

```python
#null_hdr = ['PRICE', 'YR_RMDL'] -- was going to use loop to delete rows with null values in specfic columns
dcprop_df = dcprop_df[np.isfinite(dcprop_df['PRICE'])]
dcprop_df
```

```python
#verifying that rows with null values for 'PRICE' were removed
dcprop_df.shape
```

```python
#naming column 1
dcprop_df.rename(columns={ dcprop_df.columns[0]: "ID" })

#set ID as index
#ID might not be unique
#dcprop_df.set_index('ID', inplace=True)
```

```python
import statsmodels.formula.api as smf
result = smf.ols('PRICE ~ LANDAREA + BEDRM + HF_BATHRM + KITCHENS + FIREPLACES + STYLE + GBA + EYB + AYB + YR_RMDL + GRADE + SOURCE + ZIPCODE - 1', data=dcprop_df).fit()
result.summary()
```
